```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

%matplotlib inline
```

```python
#downloading the datasets

athletes =pd.read_csv('athlete_events.csv')
region =pd.read_csv('noc_regions.csv')
```

```python
athletes.head()
```

```python
region.head()
```

```python
#join the dataframes

athletes_df = athletes.merge(region,how='left', on ='NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
athletes_df.rename(columns={'region':'Region','notes':'Notes'} , inplace = True);
```

```python
athletes_df.head()
```

```python
#Top countires participating

top_10_countries = athletes_df.Team.value_counts().sort_values(ascending = False).head(10)
top_10_countries
```

```python
#Plot for the top 10 countries

plt.figure(figsize=(12,8))
plt.title('Overall Participation by Country')
sns.barplot(x=top_10_countries.index , y= top_10_countries ,palette ='viridis');
```

```python
plt.figure(figsize=(12,8))
plt.title("Age distribuation of the athletes")
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age , bins = np.arange(10,80,2) , color='black',edgecolor='white');
```

```python
#Winter olympics sports

winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
#summer olympics sports

summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
# Male and Female participant 

gender_counts =athletes_df.Sex.value_counts()
gender_counts

```

```python

plt.figure(figsize=(12,6))
plt.title('Gender Distribuation')
plt.pie(gender_counts, labels= gender_counts.index , autopct = '%1.1f%%' , startangle= 180 , shadow=False)
plt.legend(["Male", "Fmale"]);
```

```python
#Total modals

athletes_df.Medal.value_counts()
```

```python
#Total number of femal athletes in each olympic.

female_participants = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season == 'Summer')][['Sex','Year']]

female_participants = female_participants.groupby('Year').count().reset_index()

female_participants.tail()
```

```python
womenOlympic = athletes_df[(athletes_df.Sex =='F') & (athletes_df.Season =='Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year' , data=womenOlympic ,palette ='viridis')
plt.title('Women Participation')
```

```python
part = womenOlympic.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Femal Athletes over time')
```

```python
#Gold medals athletes

goldMedals = athletes_df[(athletes_df.Medal=='Gold')]
goldMedals.head()
```

```python
#Take only the values that are different from NaN 

goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
#Gold beyond 60

goldMedals['ID'][goldMedals['Age']>60].count()
```

```python
sporting_event =goldMedals['Sport'][goldMedals['Age']>60]
sporting_event
```

```python
#Plot for sporting_event

plt.figure(figsize=(12,8))
plt.tight_layout()
sns.countplot(sporting_event ,palette ='viridis')
plt.title('Gold Medals for Athletes over 60 years')
```

```python
#Gold medals from each country

goldMedals.Region.value_counts().reset_index(name='Medal').head(5)
```

```python
totalGoldMedals = goldMedals.Region.value_counts().reset_index(name='Medal').head(6)
g =sns.catplot(x="index" ,y="Medal" ,data=totalGoldMedals ,height=5 , kind= "bar" , palette ='Blues_d')
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title('Gold Medals per Country')
```

```python
#Rio olympics

max_year =athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal =='Gold')].Team

team_names.value_counts().head(10)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=team_names.value_counts().head(20),y=team_names.value_counts().head(20).index,palette ='viridis')
plt.ylabel(None);
plt.xlabel('Countrywise Medals for the year 2016');
```
